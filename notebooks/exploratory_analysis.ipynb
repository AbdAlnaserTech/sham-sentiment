# Exploratory Data Analysis
Multilingual Sentiment Dataset — training + validation

```python
import os, sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ROOT = os.path.abspath('..')
SRC = os.path.join(ROOT, 'src')
sys.path.insert(0, SRC)

train = pd.read_csv(os.path.join(ROOT, 'data', 'sentiment_dataset_multilingual.csv'))
valid = pd.read_csv(os.path.join(ROOT, 'data', 'real', 'validation_comments.csv'))
print(len(train), len(valid))
train.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.countplot(data=train, x='language', hue='sentiment', ax=axes[0])
axes[0].set_title('Training distribution')
sns.countplot(data=valid, x='language', hue='sentiment', ax=axes[1])
axes[1].set_title('Validation distribution')
plt.tight_layout()
plt.show()
```
